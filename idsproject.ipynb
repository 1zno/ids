```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy
import random
import math
import xlrd
```

```python
#Reading Data
file = xlrd.open_workbook_xls("datafile.xls", ignore_workbook_corruption=True)
df = pd.read_excel(file)
df
```

```python
df.info() #info about the data
```

```python
p = list(df.columns)
p = p [2:14]
p
```

```python
for i in p:
    df[i] = df.groupby("SUBDIVISION")[i].transform(lambda x:x.fillna(x.mean()))
df.info()
```

```python
months =['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
```

```python
df['ANNUAL_1'] = 0
for i in months:
    df['ANNUAL_1'] = df['ANNUAL_1'] + df[i]
df.drop(columns=['ANNUAL'], inplace=True)
df.head()
```

```python
df['JF_1'] = 0
for i in ['JAN', 'FEB']:
    df['JF_1'] = df['JF_1'] + df [i]
```

```python
df['MAM_1'] = 0
for i in ['MAR', 'APR', 'MAY']:
    df['MAM_1'] = df['MAM_1'] + df [i]
```

```python
df['JJAS_1'] = 0
for i in ['JUN', 'JUL', 'AUG', 'SEP']:
    df['JJAS_1'] = df['JJAS_1'] + df [i]
```

```python
df['OND_1'] = 0
for i in ['OCT', 'NOV', 'DEC']:
    df['OND_1'] = df['OND_1'] + df [i]
```

```python
df.drop(columns=['JF', 'MAM', 'JJAS', 'OND'], inplace=True)
```

```python
df.rename(columns = {"ANNUAL_1": "ANNUAL", "JF_1": "JF", "MAM_1":"MAM", "JJAS_1": "JJAS", "OND_1": "OND"}, inplace = True)
```

```python
pop_df=df[df['SUBDIVISION']=='West Rajasthan']
pop_df.head()
```

```python
sample=pop_df[-20:]
sample.head()
```

```python
pop_annual_mean = pop_df['ANNUAL'].mean()
pop_annual_sd = pop_df['ANNUAL'].std()
sample_size = sample.shape[0]
sample_mean = sample['ANNUAL'].mean()
```

```python
alpha=0.05 #confidence is 95%
```

```python
z_stat=(sample_mean-310)/(pop_annual_sd/math.sqrt(sample_size))
z_stat
```

```python
z_alpha = 1.6449
#claim: population mean < 310
#opposite: popilation mean >= 310
#H0: population mean < 310
#H1: popilation mean >= 310
if z_stat<z_alpha:
    print('H0 is not rejected')
    print('We have enough proof to support our claim that population mean is less than 310')
else:
    print('H0 is rejected')
    print('We dont have enough proof to support our claim that population mean is less than 310')
```

```python
wr_df=df[df['SUBDIVISION']=='West Rajasthan']
first_58_yrs = wr_df.iloc[:57]
first_58_yrs.head()
```

```python
next_58_yrs = wr_df.iloc[58:116]
next_58_yrs.head()
```

```python
annual_first_58_yrs = first_58_yrs['ANNUAL'].values.tolist()
sample_annual_first_58_yrs=random.sample(annual_first_58_yrs,10)
sample_annual_first_58_yrs
```

```python
annual_next_58_yrs = next_58_yrs['ANNUAL'].values.tolist()
sample_annual_next_58_yrs = random.sample(annual_next_58_yrs, 10)
sample_annual_next_58_yrs
```

```python
f_stat,p_value=scipy.stats.f_oneway(sample_annual_first_58_yrs,sample_annual_next_58_yrs)
print('F-stat value',f_stat)
```

```python
df1=2-1
df2=2*10-2
print('df1 = ',df1, 'and df2 = ',df2)
```

```python
#F-critical values corresponding to alpha as 0.05, df1 as 1 and df2 as 18
f_critical = 4.41387331
f_critical
```

```python
if f_stat>f_critical:
    print('H0 is rejected')
    print('We dont have enough proof to support our claim that avg rainfall from 1901-1958 is equal to the avg rainfaill from 1959-2016')
else:
    print('H0 is not rejected')
    print('We have enough proof to support our claim that avg rainfall from 1901-1958 is equal to the avg rainfaill from 1959-2016')
```
